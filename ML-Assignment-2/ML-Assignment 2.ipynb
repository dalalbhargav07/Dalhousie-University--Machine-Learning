**1.2. Write a function named linear_regression to implement Linear Regression without using public libraries related to regression. Try different learning rates and show the results.**

```python
#Linear Regression
#lr = learning rate
#rp = repetition
#w0,w1 are weights
#L is a loss function
import numpy as np
def linear_regression(x,y,lr,rp):
    m = len(y)
    n = len(x[1,:])
    w0 = np.array([[-1]]).T
    w1 = np.array([[-1 for i in range(n)]]).reshape(n,1)
    L = np.array([])
    for i in range(rp -1):
        fn = np.dot(x,w1[:,-1].reshape(n,1)) + w0[-1]
        
        w1 = np.concatenate((w1,w1[:,-1].reshape(n,1) - ((lr/m)*(sum(np.dot((fn-y).reshape(1,m),x))))),1)
        w0 = np.append(w0,w0[-1] - ((lr/m) * sum(fn -y)))
        #loss Fn
        L = np.append(L,(sum((fn-y)**2))/(2*m))
    return (w0,w1,L)
def predict_regression(x,w0,w1,lr):
    n = len(x[1,:])
    target = np.dot(x,w1[:,-1].reshape(n,1)) + w0[-1]
    return (target)
```

### Ridge Regression

**1.4.	Modify the linear_regression function in a way that applies Ridge regression, and name them linear_regression_Ridge. Then repeat the assignments 1.2 and 1.3. You can thereby use a fixed learning rate that you find appropriate, but you should try and plot different values for the regularization penalty alpha.**

```python
def linear_regression_ridge(x,y,lr,lamb,rp):
    m = len(y)
    n = len(x[1,:])
    w0 = np.array([[-1]]).T
    w1 = np.array([[-1 for i in range(n)]]).reshape(n,1)
    L = np.array([])
    for i in range(rp -1):
        fn = np.dot(x,w1[:,-1].reshape(n,1)) + w0[-1]
        w1 = np.concatenate((w1,w1[:,-1].reshape(n,1) - ((lr/m)*(sum(np.dot((fn-y).reshape(1,m),x))))),1)
        w0 = np.append(w0,w0[-1] - ((lr/m) * sum(fn -y)))
        #loss Fn
        d = (sum((w1[:,-1]**2))) + (w0[-1]**2)
        L = np.append(L,(sum((fn-y)**2))/(2*m) + ((lamb/(2*m))*d))
   
    return (w0,w1,L)

def pred_regression(x,w0,w1,lr):
    n = len(x[1,:])
    target = np.dot(x,w1[:,-1].reshape(n,1)) + w0[-1]
    return (target)
```

**1.5.	Use linear_regression_Ridge and write a function named linear_regression_Ridge_momentum in which you add a momentum term. Try different momenta and plot the learning curves with and without momentum for a fixed learning rate.**

```python
def linear_regression_ridge_momentum(x,y,lr,lb,mu,rp):
    m = len(y)
    n = len(x[1,:])
    w0 = np.array([[-1]]).T
    w1 = np.array([[-1 for i in range(n)]]).reshape(n,1)
    L = np.array([])
    lamb = lb
    for i in range(rp -1):
        fn = np.dot(x,w1[:,-1].reshape(n,1)) + w0[-1]
        w1 = np.concatenate((w1,w1[:,-1].reshape(n,1) - ((lr/m)*(sum(np.dot((fn-y).reshape(1,m),x)))) + mu*(w1[:,-1] - w1[:,i-1])),1)
        w0 = np.append(w0,w0[-1] - ((lr/m) * sum(fn -y)))
        #loss Fn
        d = (sum((w1[:,-1]**2))) + (w0[-1]**2)
        L = np.append(L,(sum((fn-y)**2))/(2*m) + (lb*d/(2*m)))
   
    return (w0,w1,L)

def pred_regression_ridge(x,w0,w1,lr):
    n = len(x[1,:])
    target = np.dot(x,w1[:,-1].reshape(n,1)) + w0[-1]
    return (target)
```
